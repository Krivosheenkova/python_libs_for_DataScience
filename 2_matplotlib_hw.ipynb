Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. 

```python
a = np.arange(0, 12)
A = a.reshape(4, 3)
At = A.T
At
```

```python
B = np.dot(A, At)
B
```

Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?

```python
def inv(matrix):
    print(f'Размерность матрицы: {matrix.shape}')
    if matrix.shape[0] == matrix.shape[1] and np.linalg.det(matrix) != 0:
        print(np.linalg.inv(matrix))
    else:
        if np.linalg.det(matrix) == 0:
            print('Матрица вырождена, определитель равен нулю.')
        else:
            print('Матрица должна быть квадратной.')
```

```python
C = c.reshape(4, 4)
C
```

```python
D = B + (C * 10)
D
```

Вычислите определитель, ранг и обратную матрицу D_inv для D.

```python
D_det = np.linalg.det(D)
D_rank = np.linalg.matrix_rank(D)
D_inv = np.linalg.inv(D)
print(f'Определитель: {D_det}\nРанг: {D_rank}\nОбратная матрица:\n{D_inv}')
```

```python
D_inv = np.where(D_inv < 0, 0, 1)
D_inv
```
